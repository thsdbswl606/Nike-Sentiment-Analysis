```python
# Extract the number of reviews for each product url
import requests
from bs4 import BeautifulSoup

def fetch_product_reviews(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Ensure HTTP request was successful

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the summary element with the specific class
        summary_tag = soup.find('summary', class_='css-rptnlm')
        if summary_tag:
            # Locate the h3 element within the summary tag
            h3_tag = summary_tag.find('h3', class_='headline-4')
            if h3_tag and h3_tag.span:
                # Extracting the number from the format "Reviews (93)"
                review_count = h3_tag.span.text.strip().split('(')[1].split(')')[0]
                return review_count
            else:
                return "Review count format not found"
        else:
            return "Review summary tag not found"
    except requests.RequestException as e:
        return f"Failed to fetch page due to: {str(e)}"

product_reviews = {}

for url in product_urls:
    reviews = fetch_product_reviews(url)
    product_reviews[url] = reviews
    print(f"Number of reviews for {url}: {reviews}")
```

# t-test and chi square

t-test

```python
!pip install pandas scipy
from scipy.stats import shapiro, levene
import pandas as pd
from scipy.stats import ttest_ind
# Load the data from CSV files
import re

drive.mount('/content/drive')
path_to_non_sus = '/content/drive/My Drive/DS_project/non_sus_data.csv'
df_non_sus = pd.read_csv(path_to_non_sus)
path_to_sus = '/content/drive/My Drive/DS_project/sus_data.csv'
df_sus = pd.read_csv(path_to_sus)


# Function to clean and convert ratings
def clean_ratings(rating):
    if rating == "No rating available":
        return None  # Convert no ratings directly to None
    return pd.to_numeric(re.sub(r'\s*Stars?', '', rating, flags=re.I), errors='coerce')

# Apply cleaning function to both dataframes
df_non_sus['ratings'] = df_non_sus['ratings'].apply(clean_ratings)
df_sus['ratings'] = df_sus['ratings'].apply(clean_ratings)

df_non_sus.dropna(subset=['ratings'], inplace=True)
df_sus.dropna(subset=['ratings'], inplace=True)

# Perform the t-test on cleaned data
t_stat, p_value = ttest_ind(df_non_sus['ratings'], df_sus['ratings'], equal_var=True)
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Test for normality and equal variances
print("Normality Test for Non-Sustainable Ratings:", shapiro(df_non_sus['ratings']))
print("Normality Test for Sustainable Ratings:", shapiro(df_sus['ratings']))
print("Equal Variance Test:", levene(df_non_sus['ratings'], df_sus['ratings']))
```

chi square

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

drive.mount('/content/drive')
path_to_non_sus = '/content/drive/My Drive/DS_project/non_sus_data.csv'
df_non_sus = pd.read_csv(path_to_non_sus)
path_to_sus = '/content/drive/My Drive/DS_project/sus_data.csv'
df_sus = pd.read_csv(path_to_sus)


def clean_reviews(review):
    if review == "Review summary tag":
        return None
    return review

# Apply the cleaning function to the 'reviews' column
df_non_sus['reviews'] = df_non_sus['reviews'].apply(clean_reviews)
df_sus['reviews'] = df_sus['reviews'].apply(clean_reviews)

# Remove rows with NaN values in the 'reviews' column after cleaning
df_non_sus.dropna(subset=['reviews'], inplace=True)
df_sus.dropna(subset=['reviews'], inplace=True)

# Convert 'reviews' to numeric
df_non_sus['reviews'] = pd.to_numeric(df_non_sus['reviews'], errors='coerce')
df_sus['reviews'] = pd.to_numeric(df_sus['reviews'], errors='coerce')

# Drop any further NaN values
df_non_sus.dropna(subset=['reviews'], inplace=True)
df_sus.dropna(subset=['reviews'], inplace=True)

# Define bins for categorizing the review counts
bins = [0, 10, 50, 100, float('inf')]
labels = ['1-10', '11-50', '51-100', '101+']

# Categorize the review counts into bins
df_non_sus['review_category'] = pd.cut(df_non_sus['reviews'], bins=bins, labels=labels)
df_sus['review_category'] = pd.cut(df_sus['reviews'], bins=bins, labels=labels)

# Create a contingency table for Chi-square test
table = pd.crosstab(index=df_non_sus['review_category'], columns='Non-Sus')
table_sus = pd.crosstab(index=df_sus['review_category'], columns='Sus')
table = table.join(table_sus, how='outer').fillna(0)  # Combine and fill missing values with 0

# Perform the Chi-square test on the combined contingency table
chi2, p_value, dof, expected = chi2_contingency(table)

print("Chi-square Statistic:", chi2)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```
